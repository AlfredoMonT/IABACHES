**Proyecto:** Detección de Baches y Rejillas | **UAC**

Objetivo: Explorar el dataset, entender su estructura y distribución de clases.

```python
import os, cv2, numpy as np, pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
print('✅ Librerías OK | OpenCV:', cv2.__version__)
```

```python
PROJECT_ROOT = Path('..').resolve()
DATA_RAW     = PROJECT_ROOT / 'data' / 'raw'
FIGURES_DIR  = PROJECT_ROOT / 'results' / 'figures'
FIGURES_DIR.mkdir(parents=True, exist_ok=True)
print('Raíz:', PROJECT_ROOT)
```

```python
# Contar imágenes por split
for split in ['train', 'val', 'test']:
    p = DATA_RAW / 'images' / split
    n = len(list(p.glob('*.jpg')) + list(p.glob('*.png'))) if p.exists() else 0
    print(f'{split:6s}: {n:,} imágenes')
```

```python
# Mostrar muestra de imágenes
imgs = list((DATA_RAW / 'images' / 'train').glob('*.jpg'))[:9]
if imgs:
    fig, axes = plt.subplots(3, 3, figsize=(12,10))
    for ax, p in zip(axes.flatten(), imgs):
        img = cv2.cvtColor(cv2.imread(str(p)), cv2.COLOR_BGR2RGB)
        ax.imshow(img); ax.axis('off'); ax.set_title(p.name[:15], fontsize=8)
    plt.suptitle('Muestra — Dataset de Baches', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.savefig(FIGURES_DIR / 'muestra_imagenes.png', dpi=150)
    plt.show()
else:
    print('⚠️  Coloca el dataset en data/raw/')
```

```python
# Analizar etiquetas YOLO
labels_path = DATA_RAW / 'labels' / 'train'
anotaciones = []
if labels_path.exists():
    for f in labels_path.glob('*.txt'):
        for line in open(f).readlines():
            p = line.strip().split()
            if len(p)==5: anotaciones.append({'clase':int(p[0]),'w':float(p[3]),'h':float(p[4])})
    df = pd.DataFrame(anotaciones)
    print(f'Total anotaciones: {len(df):,}')
    print(df['clase'].value_counts())
else:
    print('⚠️  Sin carpeta de etiquetas')
```
